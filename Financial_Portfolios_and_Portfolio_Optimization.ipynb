<h3> Creating an equal-weighted Portfolio </h3>

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt 
pd.options.display.float_format = '{:.2f}'.format
```

```python
stocks = pd.read_csv("port_stocks.csv", parse_dates = ["Date"], index_col = ["Date"])
stocks
```

```python
ret = stocks.pct_change().dropna()
ret
```

```python
no_assets = len(stocks.columns)
no_assets                        # number of asssets
```

```python
weights = [1/no_assets  for i in range(0, no_assets)]     # weights for Equal-Weighted Portfolio -> 1/(no. of assets)
weights
```

```python
ret["EWP"] = ret.dot(weights)                             # or ret.mul(weights, axis = "columns").sum(axis = 1) will give the same result
ret
```

```python
summary = ret.agg(["mean", "std"]).T
summary
```

```python
summary.columns = ["Return", "Risk"]
summary
```

```python
summary.Return = summary.Return * 252
summary.Risk = summary.Risk * np.sqrt(252)

summary
```

```python
summary.plot(kind = "scatter", x = "Risk", y = "Return", s = 50, figsize = (13,9), fontsize = 15)
for i in summary.index:
  plt.annotate(i, xy = (summary.loc[i, "Risk"] + 0.002, summary.loc[i, "Return"] + 0.002))
plt.title("Risk/Return for Major Tech Stocks", fontsize = 20)
plt.xlabel("ann. Risk(std.)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.show()
```

```python
def ann_Risk_Return(returns_df):
    summary = returns_df.agg(["mean" , "std"]).T
    summary.columns = ["Return" , "Risk"]
    summary["Return"] = summary["Return"]*252
    summary["Risk"] = summary["Risk"]*np.sqrt(252)
    return summary
```

## What is the Sharpe Ratio?

\begin{equation}
Sharpe Ratio = \frac{Port Return - RFR}{Port Risk}
\end{equation}

- Port Return -> Portfolio Return
- Port Risk -> Portfolio Risk
- RFR -> Risk Free Return

```python
risk_free_return = 0.03
risk_free_risk = 0.0
```

<h3> Markowitz Portfolio Optimization </h3>

- Followed up code to plot the Efficient Frontier amongst the random portfolio for optimization

```python
print("Max Sharpe Ratio in the Array : {}".format(port_summary["Sharpe"].max()))
print("It's location in the Array : {}".format(port_summary["Sharpe"].argmax()))
```

```python
max_sr_return = port_summary.loc[port_summary.Sharpe.argmax() , "Return"]
max_sr_risk = port_summary.loc[port_summary.Sharpe.argmax(), "Risk"]

print(max_sr_return)
print(max_sr_risk)
```

```python
def get_return_sharpe_risk(sp_weights):               # this function gives ann. Return, Sharpe Ration & ann. Risk provided the weights assigned to each stock
  sp_weights = np.array(sp_weights)

  sp_ret = np.sum(ret.mean() * sp_weights * 252)
  sp_vol = np.sqrt(np.dot(sp_weights.T, np.dot(ret.cov()*252, sp_weights)))
  sp_sharpe_ratio = sp_ret/sp_vol

  return np.array([sp_ret, sp_vol, sp_sharpe_ratio])
```

```python
def neg_sharpe(sp_weights):                                 # will return negative sharpe ratio for given weight i.e Sharpe ratio * -1
  return get_return_sharpe_risk(sp_weights)[2]*-1

def check_sum(sp_weights):                                  # returns 0 if the sum of weights = 1
  return np.sum(sp_weights)-1
```

- We'll also define an initial guess and specific bounds, to help the minimization be faster and more efficient. 
- Our initial guess will be 25% for each stock (or 0.25), and the bounds will be a tuple (0,1) for each stock, since the weight can range from 0 to 1.

```python
cons = ({'type':'eq', 'fun': check_sum})
bnds = tuple((0,1) for x in range(noa))
init_guess = np.full(noa, 0.25)
init_guess
```

```python
import scipy.optimize as sco
```

```python
opt_weights = sco.minimize(neg_sharpe, init_guess, method = "SLSQP", bounds = bnds, constraints = cons)
opt_weights
```

```python
get_return_sharpe_risk(opt_weights.x)
```

```python
frontier_y = np.linspace(0.08, 0.35, 250)
```

```python
def minimum_volatility(sp_weights):
  return get_return_sharpe_risk(sp_weights)[1]
```

```python
frontier_x = []

for possible_return in frontier_y: 
  cons = ({'type':'eq', 'fun': check_sum}, 
          {'type':'eq', 'fun': lambda w: get_return_sharpe_risk(w)[0] - possible_return})
  
  result = sco.minimize(minimum_volatility, init_guess, method = 'SLSQP', bounds = bnds, constraints = cons)
  frontier_x.append(result['fun'])
```

```python
plt.figure(figsize = (15,8))
plt.scatter(port_summary.loc[:, "Risk"], port_summary.loc[:, "Return"], s = 20, 
            c = port_summary.loc[:, "Sharpe"], cmap = "viridis", alpha = 0.8)
plt.colorbar()
plt.scatter(summary.loc[:, "Risk"], summary.loc[:, "Return"], s = 40, 
            c = "black", marker = 'D')
plt.plot(frontier_x, frontier_y, 'r--', lw = 2)
plt.xticks(np.arange(0, 0.300, step = 0.025))
plt.xlabel("Ann. Risk(std.)", fontsize = 15)
plt.ylabel("Ann. Return", fontsize = 15)
plt.title("Sharpe Ratio", fontsize = 20)
plt.show()
```
